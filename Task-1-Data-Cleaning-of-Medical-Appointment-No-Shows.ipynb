```python
import pandas as pd
```

```python
df = pd.read_csv('Task 1-data-cleaning-of-Medical-Appointment-No-Shows/Medical Appointment No Shows.csv', parse_dates=['ScheduledDay','AppointmentDay'])
```

```python
print(df.shape)
print(df.dtypes)
print(df['No-show'].value_counts())
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df['Gender'] = df['Gender'].astype(str).str.strip().str.upper().replace({'F':'Female','M':'Male'})
```

```python
df
```

```python
df.loc[df['Age'] < 0, 'Age'] = pd.NA
```

```python
df
```

```python
df['wait_days'] = (df['AppointmentDay'].dt.normalize() - df['ScheduledDay'].dt.normalize()).dt.days
```

```python
df
```

```python
df['no_show'] = df['No-show'].map({'No': 0, 'Yes': 1})
```

```python
df
```

```python
df['PatientId'] = df['PatientId'].astype(str)
df['AppointmentID'] = df['AppointmentID'].astype(str)
```

```python
df
```

```python
df.to_csv('Task 1-data-cleaning-of-Medical-Appointment-No-Shows/Medical Appointment No Shows- cleaned.csv', index=False)
```
